```python
import pandas as pd
```

```python
standings_2015 = pd.read_csv('../data/seasons/2015standings.csv')
```

```python
standings_2015.columns = [u'Rk', u'Tm', u'Lg', u'G', u'W', u'L', u'W-L%', u'R', u'RA', u'Rdiff',
       u'SOS', u'SRS', u'pythWL', u'Luck', u'Inter', u'Home', u'Road',
       u'ExInn', u'1Run', u'vRHP', u'vLHP', u'vs_over_500', u'vs_under_500']
```

```python
standings_2015.head()
```

```python
standings_2015 = standings_2015.drop(standings_2015.columns[[0, 2, 3, 10, 11, 12, 13, 14, 
                                                             17, 18, 19, 20]], axis=1)
```

```python
standings_2015.head()
```

```python
standings_2015.dtypes
```

```python
standings_2015.rename(columns={'Tm': 'team', 'W': 'wins', 'L': 'losses', 'W-L%': 'win_loss', 'R': 'runs',
                               'RA': 'runs_allowed', 'Rdiff': 'run_differential','Home': 'home_success',
                               'Road': 'road_success'}, inplace=True)
```

```python
standings_2015.head()
```

```python
standings_2015['home_success'] = standings_2015.home_success.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings_2015['road_success'] = standings_2015.road_success.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings_2015['vs_over_500'] = standings_2015.vs_over_500.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
standings_2015['vs_under_500'] = standings_2015.vs_under_500.apply(lambda x: int(x.split('-')[0]) - int(x.split('-')[1]))
```

```python
standings_2015
```

```python
standings_2015.replace('CWS', 'CHW', inplace=True)
```

```python
teamzipper = dict(zip(['BOS', 'ARI', 'BAL', 'PHI', 'NYM', 'CHC', 'MIN', 'ATL', 'SFG', 'TEX', 'SDP', 'OAK','TBR', 
                'COL', 'WSN','KCR', 'TOR', 'NYY', 'MIA', 'DET', 'STL','PIT', 'CHW', 'MIL', 'CIN', 'HOU', 
                'CLE', 'LAD', 'LAA','SEA'], range(0,31)))
```

```python
for row in ["team"]:
    standings_2015[row] = standings_2015[row].apply(lambda x: teamzipper[x])
```

```python
standings_2015.dtypes
```

```python
standings_2015.to_csv('standings_2015.csv', sep='\t', encoding='utf-8')
```
