```python
def drop_and_rename_cols(df):
    df.drop(['Rk', 'Rank', 'Win', 'Loss', 'Save', 'Inn', 'Unnamed: 3'], axis=1, inplace=True)
    
    df.rename(columns={'Gm#': 'game_num', 'Date': 'date', 'Tm': 'team', 'Unnamed: 5': 'home_game', 'Opp': 'opp',
                  'W/L': 'win', 'R': 'runs_scored', 'RA': 'runs_against', 'W-L': 'record', 'GB': 'games_back',
                  'Time': 'game_time', 'D/N': 'night_game', 'Attendance': 'attendance', 'Streak': 'streak'
                  }, inplace=True)
    return df
```

```python
single_dfs = glob.glob("*.csv")
df_list = []
for filename in sorted(single_dfs):
    df = pd.read_csv(filename)
    df = drop_and_rename_cols(df)
    df_list.append(df)
full_df = pd.concat(df_list, axis=0)

full_df.to_csv('games_2015.csv')
```
