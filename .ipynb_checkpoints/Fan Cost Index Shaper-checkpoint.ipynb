```python
import pandas as pd
```

```python
fci_2015 = pd.read_csv('../rawfiles/fcis/mlb_fci_2015.csv')
```

```python
fci_2015.head()
```

```python
fci_2015.rename(columns={'Team': 'team', 'Avg. Ticket': 'avg_ticket', 'Pct. Change': 'pct_change',
                                 'Avg. Premium Ticket': 'avg_prem_ticket', 'Beer': 'beer', 'Beer Oz': 'beer_oz',
                         'Soft Drink': 'soft_drink','Soft Drink Oz.': 'soft_drink_oz', 'Hot Dog': 'hot_dog',
                         'Parking': 'parking', 'Cap': 'hat', 'FCI': 'fci', 'Change': 'fci_pct_change'},
                         inplace=True)
```

```python
fci_2015.dtypes
```

```python
fci_2015[fci_2015.team == 'Minnesota Twins']
```

```python
fci_2015.team.value_counts()
```

```python
fci_2015['team'] = fci_2015['team'].str.replace('\r', '')
```

```python
fci_2015['team'].value_counts()
```

```python
fci_2015.replace({'team' : { 'Boston Red Sox' : 'BOS', 'Arizona Diamondbacks' : 'ARI', 'Baltimore Orioles' : 'BAL',
                           'Philadelphia Phillies': 'PHI', 'New York Mets': 'NYM', 'Chicago Cubs': 'CHC',
                           'Minnesota Twins': 'MIN', 'Atlanta Braves': 'ATL', 'San Francisco Giants': 'SFG',
                           'Texas Rangers': 'TEX', 'San Diego Padres': 'SDP', 'Oakland Athletics': 'OAK',
                           'Tampa Bay Rays': 'TBR', 'Colorado Rockies':  'COL', 'Washington Nationals': 'WSN',
                           'Kansas City Royals': 'KCR', 'Toronto Blue Jays': 'TOR', 'New York Yankees': 'NYY',
                           'Miami Marlins': 'MIA', 'Detroit Tigers': 'DET', 'St. Louis Cardinals': 'STL',
                           'Pittsburgh Pirates': 'PIT', 'Chicago White Sox': 'CHW', 'Milwaukee Brewers': 'MIL',
                           'Cincinnati Reds': 'CIN', 'Houston Astros': 'HOU', 'Cleveland Indians': 'CLE',
                            'Los Angeles Dodgers': 'LAD', 'Los Angeles Angels': 'LAA', 'Seattle Mariners': 'SEA'}}, inplace=True)
```

```python
fci_2015
```

```python
teamzipper = dict(zip(['BOS', 'ARI', 'BAL', 'PHI', 'NYM', 'CHC', 'MIN', 'ATL', 'SFG', 'TEX', 'SDP', 'OAK','TBR', 
                'COL', 'WSN','KCR', 'TOR', 'NYY', 'MIA', 'DET', 'STL','PIT', 'CHW', 'MIL', 'CIN', 'HOU', 
                'CLE', 'LAD', 'LAA','SEA'], range(0,31)))
```

```python
teamzipper
```

```python
for row in ["team"]:
    fci_2015[row] = fci_2015[row].apply(lambda x: teamzipper[x])
```

```python
fci_2015.to_csv('fci2015.csv', sep='\t', encoding='utf-8')
```
