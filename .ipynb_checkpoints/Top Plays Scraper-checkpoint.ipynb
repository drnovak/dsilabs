```python
urls = ["http://www.baseball-reference.com" + url for url in urls]
box_scores_links = []
for url in urls:
    soup = BeautifulSoup(urllib2.urlopen(url).read(), "html5lib")
    for link in soup.findAll('a', href=True, text='Box Score and Play-By-Play'):
        box_scores_links.append('http://www.baseball-reference.com'+link['href'])
```

```python
#and pull "Top 5 Plays" table from each page

top_5_plays = {}
urls = box_scores_links
for url in urls:
    print url
    soup = BeautifulSoup(urllib2.urlopen(url).read(), "html5lib")
    plays = pd.read_html(url, attrs={"id":'top_plays'}, encoding='utf-8')
    print plays
    top_5_plays[url] = plays
    
## To access df out of top5

## top_5_plays[whatever_URL][0]
```
