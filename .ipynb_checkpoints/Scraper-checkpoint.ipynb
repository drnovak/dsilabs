```python
def get_salary_info(soup):
    player_salary_information_2016 = []
    row = soup.findAll('tr',{'class': "page"})
    items = soup.findAll({'td':'class'})
    for items in row[:]:
        player_salary_information_2016.append(items.get_text(strip=True))
    return player_salary_information_2016
    
players2016 = get_salary_info(soup)

print players2016[0:5]
```
