```python
import requests
```

```python
import base64

app_token = raw_input('App token: ')
consumer_key = raw_input('Enter consumer key: ')
consumer_secret = raw_input('Consumer Secret: ')
stubhub_user = raw_input('User email: ')
stubhub_pass = raw_input('Password: ')

```

```python
combo = consumer_key + ':' + consumer_secret
basic_auth = base64.b64encode(combo)
```

```python
headers = {
    'Content-Type': 'application/x-www-form-urlencoded',
    'Authorization': 'Basic ' + basic_auth
}
```

```python
body = {
    'grant_type': 'password',
    'username': stubhub_user,
    'password': stubhub_pass,
    'scope': 'PRODUCTION'
}
```

```python
login_request = requests.post('https://api.stubhub.com/login', headers=headers, data=body)
```

```python

login_request
```

```python
login_request.text
```

```python
access_token = login_request.json()['access_token']
```

```python
user_GUID = login_request.headers['X-StubHub-User-GUID']
user_GUID
```

```python
headers['Authorization'] = 'Bearer ' + access_token
headers['Accept'] = 'application/json'
headers['Accept-Encoding'] = 'application/json'

search_data = {
    'q': 'cubs',
    'city': 'chicago'
    }

search_url = 'https://api.stubhub.com/search/catalog/events/v3'

cubs_events = requests.get(search_url, headers=headers, params=search_data)
```
